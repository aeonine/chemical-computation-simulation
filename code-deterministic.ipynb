```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def schlogl_ode(t, x, k1, k2):
    return [k1 * x**2 - k2 * x**3 - x]
```

```python
k1 = 2
k2 = 0.4

t_span = (0.0, 10.0)
t_eval = np.linspace(*t_span, 1000)


fig1, ax1 = plt.subplots(figsize=(7, 4))

result = solve_ivp(
    schlogl_ode,
    t_span,
    [0.5635083268962915],
    args=(k1, k2),
    t_eval=t_eval,
)

ax1.plot(result.t, result.y[0])
ax1.set_xlabel("Time")
ax1.set_ylabel("Concentration of X")
ax1.set_ylim(-0.1, 4.5)

plt.tight_layout()
```

```python
def get_xhigh(k1, k2):
    return (k1 + np.sqrt(k1**2 - 4 * k2)) / (2 * k2)

def get_xcrit(k1, k2):
    return (k1 - np.sqrt(k1**2 - 4 * k2)) / (2 * k2)
```

```python
k1 = 2
k2 = 0.4

x_crit =  get_xcrit(k1, k2)
x_high =  get_xhigh(k1, k2)
x_low  = 0.0

t_span = (0.0, 10.0)
t_eval = np.linspace(*t_span, 1000)

initial_concentrations = [
    0.5  * x_crit,
    0.6  * x_crit,
    0.7  * x_crit,
    0.8  * x_crit,
    0.95  * x_crit,
    1.05  * x_crit,
    1.1  * x_crit,
    1.2  * x_crit,
    1.3  * x_crit,
    1.4  * x_crit,
]

labels = [
    "0.5","0.6","0.7","0.8","0.95",
    "1.05","1.1","1.2","1.3","1.4"
]

fig1, ax1 = plt.subplots(figsize=(7, 4))

for conc, label in zip(initial_concentrations, labels):
    result = solve_ivp(
        schlogl_ode,
        t_span,
        [conc],
        args=(k1, k2),
        t_eval=t_eval,
    )
    ax1.plot(result.t, result.y[0], label=label)

ax1.axhline(x_crit, linestyle="--",  color="k", label=r"$x_{\mathrm{crit}}$")
ax1.set_xlabel("Time")
ax1.set_ylabel("Concentration of X")
ax1.legend(title="Initial Concentrations", fontsize=8, loc="best")
ax1.set_ylim(-0.1, 4.5)
plt.tight_layout()

```
